### –ò—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –¥–∏–Ω–∞–º–∏–∫–∏ –ø—Ä–æ–¥–∞–∂ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º

### –ò—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –≤–ª–∏—è–Ω–∏—è —Ä–µ–π—Ç–∏–Ω–≥–æ–≤ –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏

–ò–Ω—Ç—É–∏—Ç–∏–≤–Ω–æ –∫–∞–∂–µ—Ç—Å—è, —á—Ç–æ –º–µ–∂–¥—É –æ—Ç–∑—ã–≤–∞–º–∏ –∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å —Å–∏–ª—å–Ω–∞—è –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è. –ò —Ö–æ—Ç—è –º—ã –Ω–µ —Å–º–æ–∂–µ–º —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥ –æ –ø—Ä–∏—á–∏–Ω–Ω–æ-—Å–ª–µ–¥—Å—Ç–≤–µ–Ω–Ω—ã—Ö —Å–≤—è–∑—è—Ö, —É—Å—Ç–∞–Ω–æ–≤–∏–º —ç—Ç—É —Å–≤—è–∑—å.

```python
for platform in actual_platforms:
    plt.figure(figsize=(16, 5)) # –∑–∞–¥–∞—ë–º —Ä–∞–∑–º–µ—Ä —Å–µ—Ç–∫–∏
    plt.suptitle('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –≤–Ω–µ—à–Ω–∏–º–∏ –æ—Ü–µ–Ω–∫–∞–º–∏ –∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏–≥—Ä –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ ' + str(platform) +'\n', y=1.05, size=14)
    gridsize = 40

    ax = plt.subplot(1,2,1)
    data_filtered.query('platform==@platform').plot(x='total_sales', 
                                                    y='critic_score', 
                                                    kind='hexbin',
                                                    gridsize = gridsize,
                                                    ax=ax);
    plt.title('–°–≤—è–∑—å –ø—Ä–æ–¥–∞–∂ —Å –æ—Ü–µ–Ω–∫–æ–π –∫—Ä–∏—Ç–∏–∫–æ–≤', pad=10);
    plt.xlabel('–ß–∏—Å–ª–æ –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–æ–ø–∏–π –∏–≥—Ä—ã, –º–ª–Ω');
    plt.ylabel('–û—Ü–µ–Ω–∫–∞ –∫—Ä–∏—Ç–∏–∫–æ–≤');
    

    ax = plt.subplot(1,2,2)
    data_filtered.query('platform==@platform').plot(x='total_sales', 
                                                    y='user_score', 
                                                    kind='hexbin',
                                                    gridsize = gridsize,
                                                    alpha=1, 
                                                    ax=ax);
    plt.title('–°–≤—è–∑—å –ø—Ä–æ–¥–∞–∂ —Å –æ—Ü–µ–Ω–∫–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π', pad=10);
    plt.xlabel('–ß–∏—Å–ª–æ –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–æ–ø–∏–π –∏–≥—Ä—ã, –º–ª–Ω');
    plt.ylabel('–û—Ü–µ–Ω–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π');
    
    critics_corr = round(data_filtered.query('platform==@platform')["total_sales"].corr(data_filtered.query('platform==@platform')["critic_score"]), 2)
    users_corr = round(data_filtered.query('platform==@platform')["total_sales"].corr(data_filtered.query('platform==@platform')["user_score"]), 2)
    
    plt.figtext(0.1, -0.05, 
    f'–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —Å –æ—Ü–µ–Ω–∫–æ–π –∫—Ä–∏—Ç–∏–∫–æ–≤: {critics_corr}')
    
    plt.figtext(0.1, -0.1, 
    f'–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —Å –æ—Ü–µ–Ω–∫–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π: {users_corr}')
    
    plt.figtext(0.1, -0.2, ' ') 
```

–û–±—Ä–∞—Ç–∏–º—Å—è –∫ –æ—Ç–¥–µ–ª—å–Ω—ã–º –∂–∞–Ω—Ä–∞–º –∏ –ø–æ—Å—Ç—Ä–æ–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–∂ –≤ —Ä–∞–∑—Ä–µ–∑–µ —ç—Ç–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞.

```python
regions_dic = {'–°–µ–≤–µ—Ä–Ω–∞—è –ê–º–µ—Ä–∏–∫–∞':'north_america_sales', '–ï–≤—Ä–æ–ø–∞': 'europe_sales', '–Ø–ø–æ–Ω–∏—è': 'japan_sales'}

for region in regions_dic:
    hhh = (data_filtered.
       query('platform in @actual_platforms and year_of_release>=2013').
       pivot_table(index='genre', 
                   columns='platform', 
                   values = regions_dic[region], 
                   aggfunc = 'median').
       fillna(0))

    plt.subplots(figsize=(16,8)) 
    plt.title('–ú–µ–¥–∏–∞–Ω–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏ –≤ —Ä–µ–≥–∏–æ–Ω–µ ' + region + ', –º–ª–Ω', y=1.02, size=14)
    sns.heatmap(hhh, annot=True);
    plt.xlabel('');
    plt.ylabel('');
```

```python
rating_to_regions = (data_filtered.pivot_table(index='rating',
                     values = ['north_america_sales', 'europe_sales', 'japan_sales'],
           aggfunc = {'north_america_sales': lambda x: (100*x.sum()/data_filtered['north_america_sales'].sum()).round(1), 
                                      'japan_sales': lambda x: (100*x.sum()/data_filtered['japan_sales'].sum()).round(1), 
                                      'europe_sales': lambda x: (100*x.sum()/data_filtered['europe_sales'].sum()).round(1)}))

rating_to_regions.rename(columns={'europe_sales': '–ü—Ä–æ–¥–∞–∂–∏ –≤ –ï–≤—Ä–æ–ø–µ, %', 
                                  'japan_sales': '–ü—Ä–æ–¥–∞–∂–∏ –≤ –Ø–ø–æ–Ω–∏–∏, %',
                                  'north_america_sales': '–ü—Ä–æ–¥–∞–∂–∏ –≤ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ, %'}, inplace=True)

rating_to_regions
```

## –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑

### –ì–∏–ø–æ—Ç–µ–∑–∞ 1

–ì–∏–ø–æ—Ç–µ–∑–∞ —Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ—Ç—Å—è —Ç–∞–∫: —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ. –ß—Ç–æ–±—ã –µ–µ –ø–æ–¥—Ç–≤–µ—Ä–¥–∏—Ç—å –∏–ª–∏ –æ–ø—Ä–æ–≤–µ—Ä–≥–Ω—É—Ç—å, –Ω—É–∂–Ω–æ —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –Ω—É–ª–µ–≤—É—é –∏ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—É—é –≥–∏–ø–æ—Ç–µ–∑—ã

* $H_0$: –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º –Ω–µ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è
* $H_1$: –º–µ–∂–¥—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–º–∏ —Ä–µ–π—Ç–∏–Ω–≥–∞–º–∏ –µ—Å—Ç—å –∑–Ω–∞—á–∏–º–æ–µ –æ—Ç–ª–∏—á–∏–µ

–ü—É—Å—Ç—å –ø–æ—Ä–æ–≥–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è –ø—Ä–∏–Ω—è—Ç–∏—è —Ä–µ—à–µ–Ω–∏—è –æ —Ü–µ–ª–µ—Å–æ–æ–±—Ä–∞–∑–Ω–æ—Å—Ç–∏ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –Ω—É–ª–µ–≤–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã –±—É–¥–µ—Ç alpha=5%

```python
pvalue = st.ttest_ind(data_filtered.query('year_of_release>=2013 and platform=="XOne" and not user_score.isna()')['user_score'], 
             data_filtered.query('year_of_release>=2013 and platform=="PC" and not user_score.isna()')['user_score'],
             equal_var = False).pvalue

if pvalue > 0.05:
    print('–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º –Ω–µ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è, p-value —Ä–∞–≤–Ω–æ', round(pvalue, 2))
else:
    print('–ú–µ–∂–¥—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–º–∏ —Ä–µ–π—Ç–∏–Ω–≥–∞–º–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º –µ—Å—Ç—å –∑–Ω–∞—á–∏–º–æ–µ –æ—Ç–ª–∏—á–∏–µ, p-value —Ä–∞–≤–Ω–æ', round(pvalue, 6))
```

### –ì–∏–ø–æ—Ç–µ–∑–∞ 2

–°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action (–∞–Ω–≥–ª. ¬´–¥–µ–π—Å—Ç–≤–∏–µ¬ª, —ç–∫—à–µ–Ω-–∏–≥—Ä—ã) –∏ Sports (–∞–Ω–≥–ª. ¬´—Å–ø–æ—Ä—Ç–∏–≤–Ω—ã–µ —Å–æ—Ä–µ–≤–Ω–æ–≤–∞–Ω–∏—è¬ª) —Ä–∞–∑–Ω—ã–µ.

–ö–∞–∫–æ–π –±—ã –Ω–∏ –±—ã–ª–∞ —Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∫–∞ –≤—ã–¥–≤–∏–Ω—É—Ç–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã, –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞, –∫–∞–∫ –ø—Ä–∞–≤–∏–ª–æ, –≥–æ–≤–æ—Ä–∏—Ç –æ —Ç–æ–º, —á—Ç–æ –Ω–∞ —Å–∞–º–æ–º –¥–µ–ª–µ —Ä–∞–∑–ª–∏—á–∏–π –º–µ–∂–¥—É –≤—ã–±–æ—Ä–∫–∞–º–∏ –Ω–µ—Ç.

* $H_0$: –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports –Ω–µ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è
* $H_1$: –º–µ–∂–¥—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–º–∏ —Ä–µ–π—Ç–∏–Ω–≥–∞–º–∏ Action –∏ Sports –µ—Å—Ç—å –æ—Ç–ª–∏—á–∏–µ

```python
pvalue = st.ttest_ind(data_filtered.query('year_of_release>=2013 and genre=="Action" and not user_score.isna()')['user_score'], 
             data_filtered.query('year_of_release>=2013 and genre=="Sports" and not user_score.isna()')['user_score'],
             equal_var = False).pvalue

if pvalue > 0.05:
    print('–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports –Ω–µ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è, p-value —Ä–∞–≤–Ω–æ', round(pvalue, 20))
else:
    print('–ú–µ–∂–¥—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–º–∏ —Ä–µ–π—Ç–∏–Ω–≥–∞–º–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports –µ—Å—Ç—å –∑–Ω–∞—á–∏–º–æ–µ –æ—Ç–ª–∏—á–∏–µ, p-value —Ä–∞–≤–Ω–æ', round(pvalue, 20))
```
